```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, limit
from IPython.display import display, Math
```

# Limites

## Definición

Cuál es el valor de `y` cuando los números se acercan a `x`

$\lim_{x \to 0} algo$

Se lee, límite cuando la `x` tiende a cero de `algo`.

$\lim_{x \to 2} (x + 1) = 2$

Límite cuando la `x` tiende a cero de $ (x + 1) $


Para calcular el límite, se reemplaza la `x` y calculamos.

$\lim_{x \to 2} (x + 1) =$

$\lim_{x \to 2} (2 + 1) = 3$ 

Otra:

$\lim_{x \to 3} (x + 1) = $

$\lim_{x \to 3} (3 + 1) = 4$

```python
x = symbols('x')
expression = x + 1

limit_value = limit(expression, x, 2)

display(Math(rf'$\lim_{{x \to 2}} (x + 1) = {limit_value}$'))
```

```python
expression = 2 * x**2

limit_value = limit(expression, x, 4)
display(Math(rf'$\lim_{{x \to -4 }} 2x^{{2}} = {limit_value}$'))
```

En un plano cartesiano, podemos calcular la 'altura', osea `y` para esta función.

```python
def f(x):
    return x + 1

x = np.linspace(-5, 5, 100)
y = f(x)

fig, ax = plt.subplots()
ax.plot(x, y)

ax.axvline(x=0, color='black', linestyle='--')  # x horizontal on zero
ax.axhline(y=0, color='black', linestyle='--')  # y vertical on zero

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Gráfica de la función f(x) = x + 2')

# horizontal line to describe de limit
ax.hlines(y=3, xmin=0, xmax=2, color='red', linestyle='--')

# vertical line to describe the limit

ax.vlines(x=2, ymin=0, ymax=3, color='red', linestyle='--')

# adjust ticks
plt.xticks(np.arange(-5, 6, step=1))
plt.yticks(np.arange(-5, 6, step=1))

plt.grid(True)
plt.show()
```

## Límite por factorización

### Diferencias de cuadrados

- Sólo se puede cuando hay dos términos
- Tienen que ser cuadrados perfectos
- Que se les puede sacar raiz cuadrada.

$\lim_{x \to 3} \frac{x^2 - 9}{x-3} $

Para calcular el límite $\lim_{x \to 3} \frac{x^2 - 9}{x-3}$, podemos simplificar la expresión factorizando el numerador y cancelando el factor común de $(x-3)$:

Dos parentesis, uno `+` y otro `-`

$\lim_{x \to 3} \frac{x^2 - 9}{x-3} = \lim_{x \to 3} \frac{(x-3)(x+3)}{x-3}$

Luego, podemos simplificar aún más, ya que $(x-3)$ se cancela en el numerador y denominador:

$\lim_{x \to 3} \frac{(x-3)(x+3)}{x-3} = \lim_{x \to 3} (x+3)$

Finalmente, evaluamos el límite cuando $x$ tiende a 3:

$\lim_{x \to 3} (x+3) = 3 + 3 = 6$

Por lo tanto, el límite de la expresión $\frac{x^2 - 9}{x-3}$ cuando $x$ tiende a 3 es igual a 6.

```python
expression = (x**2 - 9) / (x - 3)

limit_value = limit(expression, x, 3)
display(Math(rf'$\lim_{{x \to 3}} \frac{{x^2 - 9}}{{x-3}} = {limit_value}$'))
```

### Factor común

- Tiene que existir algo que se repita en los términos

$\lim_{x \to 2} \frac{x^{2} - 2x}{x - 2}$

Para calcular el límite $\lim_{x \to 2} \frac{x^2 - 2x}{x - 2}$ 

mediante el factor común, podemos simplificar la expresión y eliminar la indeterminación en el denominador. Aquí está el cálculo paso a paso:

$\lim_{x \to 2} \frac{x^2 - 2x}{x - 2}$

$=\lim_{x \to 2} \frac{x(x - 2)}{x - 2}$

$=\lim_{x \to 2} x$

Por lo tanto, el límite es simplemente $2$ cuando $x$ tiende a $2$.

```python
expression = (x**2 - 2*x) / (x - 2)

limit_value = limit(expression, x, 3)
display(Math(rf'$\lim_{{x \to 3}} \frac{{x^2 - 9}}{{x-3}} = {limit_value}$'))
```
